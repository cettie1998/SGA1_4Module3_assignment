```python
%%sql
SELECT first_name and last_name AS NAME, SALARY
FROM employees
WHERE NAME IS first_name and last_name;
```

```python
%%sql
SELECT *
FROM employees
ORDER BY SALARY ASC;
```

```python
%%sql
SELECT COUNT(*) AS "salary" 
FROM employees
WHERE salary > '200000';
```

```python
%%sql
SELECT MAX(salary) as "highest salary"
FROM employees;
```

```python
%%sql
SELECT MIN(salary) as "least salary"
FROM employees;
```
